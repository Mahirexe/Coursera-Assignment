```python
# Convert minutes to hours
minutes = 135

hours = minutes // 60  # integer division to get the whole number of hours
minutes_remainder = minutes % 60  # modulus operator to get the remaining minutes

print("The converted time is:", hours, "hours", minutes_remainder, "minutes")
```
